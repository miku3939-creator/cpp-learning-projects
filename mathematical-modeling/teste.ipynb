```python
import numpy as np
import matplotlib.pyplot as plt

# 解决中文字体显示问题
plt.rcParams['font.sans-serif'] = ['SimHei', 'Microsoft YaHei', 'Arial Unicode MS']
plt.rcParams['axes.unicode_minus'] = False

# 【表 3 实际测量与计算数据输入】
y3 = np.array([-50, -40, -30, -20, -10, 0, 10, 20, 30])
b3_meas = np.array([0.139, 0.141, 0.143, 0.143, 0.143, 0.143, 0.143, 0.143, 0.143])

# 创建画布
plt.figure(figsize=(8, 5))

# 绘制径向分布曲线
plt.plot(y3, b3_meas, 'ms-', label='径向磁场测量值 $B(Y)$', markersize=6, linewidth=1.5, color='darkmagenta')

# 绘制均匀磁场的基准参考线
plt.axhline(y=0.143, color='gray', linestyle='--', linewidth=1, alpha=0.7)

# 适度微调纵轴范围，以便清晰看清平台区的细微波动，同时不失真
plt.ylim(0.120, 0.160)

# 图表美化与标签设置
plt.title('亥姆霍兹线圈中心径向磁场分布曲线 (表3)', fontsize=13, pad=12)
plt.xlabel('径向距离 $Y$ (mm)', fontsize=11)
plt.ylabel('磁感应强度 $B$ (mT)', fontsize=11)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=10, loc='lower center')
plt.tight_layout()

# 显示图像
plt.show()
```
